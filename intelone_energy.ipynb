## Processing time  column.

### In this dataset, time is recorded in the Unix Time format. 
 - Unix Time represents the number of seconds that have passed since 00:00:00 UTC Thursday, 1 January 1970

```python
home_df['time'].head()
```

### I am converting this large number that represents a unix timestamp (i.e. "1451624400") to a readable date.

## Extrating datetime information such as year, month and day from time column.

```python

home_df['year'] = home_df['time'].apply(lambda x : x.year)
home_df['month'] = home_df['time'].apply(lambda x : x.month)
home_df['day'] = home_df['time'].apply(lambda x : x.day)
home_df['weekday'] = home_df['time'].apply(lambda x : x.day_name())
home_df['week_of_year'] = home_df['time'].apply(lambda x : x.weekofyear)
home_df['hour'] = home_df['time'].apply(lambda x : x.hour)
home_df['minute'] = home_df['time'].apply(lambda x : x.minute)


home_df.head()
```

## Timing information.

Hour variable can be broken into Night, Morning, Afternoon and Evening based on its number.
 - Night : 22:00 - 23:59 / 00:00 - 03:59
 - Morning : 04:00 - 11:59
 - Afternoon : 12:00 - 16:59
 - Evening : 17:00 - 21:59
 
We can create timing variable based on hour variable.

```python
def time_of_day(x):
    if x in [22,23,0,1,2,3]:
        time_of_day = 'Night'
    elif x in range(4, 12):
        time_of_day = 'Morning'
    elif x in range(12, 17):
        time_of_day = 'Afternoon'
    elif x in range(17, 22):
        time_of_day = 'Evening'
    else:
        time_of_day = 'X'
    return time_of_day
```

```python
home_df['time_of_day'] = home_df['hour'].apply(time_of_day)

home_df.head()
```

# Time Series Analysis

##### Defining required functions.

```python
# function for grouping a particular column by number of day.
def group_by_day(col, df):
    day_df = df.set_index(df['time'])
    day_df = day_df.resample('D').mean()
    day_df['day_num']=range(1,352)
    day_df = day_df[[col,'day_num']].groupby('day_num').agg({col:['mean']})[col]
    return day_df

# function for grouping a particular column by month.
def group_by_month(col):
    return home_df[[col,'month']].groupby('month').agg({col:['mean']})[col]

# function for grouping a particular column by weekday.
def group_by_day_of_week(col):
    weekday_df = home_df.groupby('weekday').agg({col:['mean']})
    weekday_df.columns = [f"{i[0]}_{i[1]}" for i in weekday_df.columns]
    weekday_df['week_num'] = [['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'].index(i) for i in weekday_df.index]
    weekday_df.sort_values('week_num', inplace=True)
    weekday_df.drop('week_num', axis=1, inplace=True)
    return weekday_df

# function for grouping a particular column by time of the day.
def group_by_time_of_day(col):
    timing_df = home_df.groupby('time_of_day').agg({col:['mean']})
    timing_df.columns = [f"{i[0]}_{i[1]}" for i in timing_df.columns]
    timing_df['timing_num'] = [['Morning','Afternoon','Evening','Night'].index(i) for i in timing_df.index]
    timing_df.sort_values('timing_num', inplace=True)
    timing_df.drop('timing_num', axis=1, inplace=True)
    return timing_df

# function for plotting time series.
def ts_plot(df, color, title, x_label, y_label):
    plt.figure(figsize=(15,6))
    sbn.set_theme(style="whitegrid")
    sbn.lineplot(data= df, palette=[color])
    plt.title(title)
    plt.xlabel(x_label)
    plt.ylabel(y_label)
    plt.show()
```

<h3>ARIMA Model</h3>

```python
from statsmodels.tsa.arima_model import ARIMA
from sklearn.metrics import median_absolute_error, mean_squared_error, mean_squared_log_error
def forcast_ts(data, tt_ratio):
    X = data.values
    size = int(len(X) * tt_ratio)
    train, test = X[0:size], X[size:len(X)]
    history = [x for x in train]
    predictions = list()
    for t in range(len(test)):
        model = ARIMA(history, order=(5,1,0))
        model_fit = model.fit(disp=0)
        output = model_fit.forecast()
        yhat = output[0]
        predictions.append(yhat)
        obs = test[t]
        history.append(obs)

    mae = mean_absolute_error(test, predictions)
    mse = mean_squared_error(test, predictions)

    print('\n Test MAE: %.5f' % mae)
    print('\n Test MSE: %.5f' % mse)

    plt.rcParams["figure.figsize"] = (25,10)
    preds = np.append(train, predictions)
    plt.plot(list(preds), color='green', linewidth=3, label="Predicted Data")
    plt.plot(list(data), color='blue', linewidth=2, label="Original Data")
    plt.title("Energy Consumption Prediction using ARIMA model.", fontsize=25)
    plt.axvline(x=int(len(data)*tt_ratio)-1, linewidth=5, color='red')
    plt.xlabel("Weeks", fontsize=22)
    plt.ylabel("Energy Consumption in KW", fontsize=22)
    plt.xticks(fontsize=18)
    plt.yticks(fontsize=18)
    plt.legend()
    plt.show()
    return mae, mse
```

<h3>LightGBM Regressor</h3>

```python
def hours2timing(x):
    if x in [22,23,0,1,2,3]:
        timing = 'Night'
    elif x in range(4, 12):
        timing = 'Morning'
    elif x in range(12, 17):
        timing = 'Afternoon'
    elif x in range(17, 22):
        timing = 'Evening'
    else:
        timing = 'X'
    return timing

home_df['timing'] = home_df['hour'].apply(hours2timing)
home_df.head(3)
```

```python
from sklearn.preprocessing import LabelEncoder
import lightgbm as lgb
```

```python
_lgbm_df = home_df.resample('w').mean()
_lgbm_df['weekday'] =   LabelEncoder().fit_transform(pd.Series(_lgbm_df.index).apply(lambda x : x.day_name())).astype(np.int8)
_lgbm_df['timing'] = LabelEncoder().fit_transform(_lgbm_df['hour'].apply(hours2timing)).astype(np.int8)
```

```python
def lgbm_train(cols=['temperature','dewPoint','use_HO'],trg='use_HO',train_ratio=0.8,valid_ratio=0.1,test_ratio=0.1):
    #make dataframe for training
    lgbm_df = _lgbm_df[cols]
    tr,vd,te = [int(len(lgbm_df) * i) for i in [train_ratio, valid_ratio, test_ratio]]
    X_train, Y_train = lgbm_df[0:tr].drop([trg], axis=1), lgbm_df[0:tr][trg]
    X_valid, Y_valid = lgbm_df[tr:tr+vd].drop([trg], axis=1), lgbm_df[tr:tr+vd][trg]
    X_test = lgbm_df[tr+vd:tr+vd+te+2].drop([trg], axis=1)
    lgb_train = lgb.Dataset(X_train, Y_train)
    lgb_valid = lgb.Dataset(X_valid, Y_valid, reference=lgb_train)
    #model training
    params = {
        'task' : 'train',
        'boosting':'gbdt',
        'objective' : 'regression',
        'metric' : {'mse'},
        'num_leaves':300,
        'drop_rate':0.05,
        'learning_rate':0.3,
        'min_child_samples':5,
    }
    gbm = lgb.train(params, lgb_train, num_boost_round=100, valid_sets=[lgb_train, lgb_valid], early_stopping_rounds=100)
    
    #make predict dataframe
    pre_df = pd.DataFrame()
    pre_df[trg] = gbm.predict(X_test, num_iteration=gbm.best_iteration)
    pre_df.index = lgbm_df.index[tr+vd:tr+vd+te+2]
    return pre_df, gbm, X_train
```
